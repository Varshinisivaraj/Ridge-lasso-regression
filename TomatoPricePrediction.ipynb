```python
Kerala = df.query("state == 'Kerala'")
ala=Kerala.query("market == 'Alappuzha'")
ala.shape
```

```python
ala.head()
```

```python
# Create lag features for the AR model
p=2
for lag in range(1, p + 1):
    ala[f'Lag_{lag}'] = ala['modal_price'].shift(lag)
```

```python
# Split the data into training and test sets
train_size = int(0.8 * len(ala))
train_data = ala[:train_size]
test_data = ala[train_size:]
```

```python
import statsmodels.api as sm
# Fit the AR model
X_train = train_data[['Lag_' + str(lag) for lag in range(1, p + 1)]]
y_train = train_data['modal_price']

# First, remove rows with NaN or Inf values from both X_train and y_train
X_train = X_train.dropna()
X_train = X_train[~X_train.isin([np.nan, np.inf, -np.inf]).any(1)]
y_train = y_train[X_train.index]  # Align the index of y_train with X_train

# Now, you can proceed with model fitting
model = sm.OLS(y_train, sm.add_constant(X_train)).fit()
```

```python
Kerala.head()
```

```python
# Print the model summary
print(model.summary())
```

```python
# Make predictions on the test data
X_test = test_data[['Lag_' + str(lag) for lag in range(1, p + 1)]]
X_test = sm.add_constant(X_test)  # Add a constant term for the intercept
y_pred = model.predict(X_test)
```

```python
# Plot the observed vs. predicted values
plt.figure(figsize=(12, 6))
plt.plot(test_data['arrival_date'], test_data['modal_price'], label='Observed', marker='o')
plt.plot(test_data['arrival_date'], y_pred, label='Predicted', marker='x')
plt.xlabel('Time')
plt.ylabel('Value')
plt.title('AR Model Forecasting')
plt.legend()
plt.show()
```

```python
# Assuming model is your trained AR model
n_forecast = 10  # Number of future time periods to forecast
forecasted_values = []

# Start with the last known values from your historical data
last_known_values = X_train.iloc[-1].values.tolist()

for _ in range(n_forecast):
    # Use the AR model to predict the next value
    next_value = model.predict([1] + last_known_values)  # 1 is the constant term
    forecasted_values.append(next_value[0])

    # Update the last_known_values list for the next iteration
    last_known_values = [next_value[0]] + last_known_values[:-1]

# The forecasted_values list now contains your future price predictions
print(forecasted_values)
```
